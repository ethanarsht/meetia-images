```python
import pandas as pd
import numpy as np
from bokeh.charts.attributes import cat
from bokeh.models import HoverTool
from bokeh.plotting import ColumnDataSource, figure, show, output_file, save
from bokeh.models.formatters import BasicTickFormatter
import bokeh.palettes
from bokeh.models.widgets import DataTable, DateFormatter, TableColumn
from bokeh.layouts import gridplot
import math
```

```python
df_cvent = pd.read_excel('cvent_50_lists.xlsx')
```

```python
df_cvent.head()
```

```python
city_list = list(pd.concat([df_cvent[2012], df_cvent[2013], df_cvent[2014], df_cvent[2015], df_cvent[2016], df_cvent[2017]]).unique())
```

```python
for i in df_cvent[2012]:
    print i
```

```python
list_12 = []
list_13 = []
list_14 = []
list_15 = []
list_16 = []
list_17 = []
for c in city_list:
    df_c = df_cvent[df_cvent[2012] == c]
    if df_c.empty == False:
        df_c = df_c.reset_index()
        rank = df_c['index'].loc[0] + 1
        list_12.append({2012: rank,
                         'city': c})
    else:
        rank = 'unranked'
        list_12.append({2012: rank,
                        'city':c})
    df_2012 = pd.DataFrame(list_12)
    
    
    df_c = df_cvent[df_cvent[2013]== c]
    if df_c.empty == False:
        df_c = df_c.reset_index()
        rank = df_c['index'].loc[0] + 1
        list_13.append({2013: rank,
                         'city':c})
    else:
        rank = 'unranked'
        list_13.append({2013: rank,
                         'city': c})
    df_2013 = pd.DataFrame(list_13)
    
    
    df_c = df_cvent[df_cvent[2014]== c]
    if df_c.empty == False:
        df_c = df_c.reset_index()
        rank = df_c['index'].loc[0] + 1
        list_14.append({2014: rank,
                         'city':c})
    else:
        rank = 'unranked'
        list_14.append({2014: rank,
                         'city': c})
    df_2014 = pd.DataFrame(list_14)
    
    
    df_c = df_cvent[df_cvent[2015]== c]
    if df_c.empty == False:
        df_c = df_c.reset_index()
        rank = df_c['index'].loc[0] + 1
        list_15.append({2015: rank,
                         'city':c})
    else:
        rank = 'unranked'
        list_15.append({2015: rank,
                         'city': c})
    df_2015 = pd.DataFrame(list_15)
    
    
    df_c = df_cvent[df_cvent[2016]== c]
    if df_c.empty == False:
        df_c = df_c.reset_index()
        rank = df_c['index'].loc[0] + 1
        list_16.append({2016: rank,
                         'city':c})
    else:
        rank = 'unranked'
        list_16.append({2016: rank,
                         'city': c})
    df_2016 = pd.DataFrame(list_16)
    
    
    df_c = df_cvent[df_cvent[2017]== c]
    if df_c.empty == False:
        df_c = df_c.reset_index()
        rank = df_c['index'].loc[0] + 1
        list_17.append({2017: rank,
                         'city':c})
    else:
        rank = 'unranked'
        list_17.append({2017: rank,
                         'city': c})
    df_2017 = pd.DataFrame(list_17)
        
```

```python
df_ranks = df_2012.merge(df_2013, on = 'city').merge(df_2014, on = 'city').merge(df_2015, on = 'city').merge(df_2016, on = 'city').merge(df_2017, on = 'city')
```

```python
df_bar.columns = ['city', '2012', '2013', '2014', '2015','2016', '2017', 'change', '2012_hover', '2013_hover', '2014_hover', '2015_hover', '2016_hover', '2017_hover']
```

```python
df_bar = df_bar.sort('change', ascending = True)
```

```python
df_bar['percent_change'] = (df_bar['change'] / df_bar['2012']) * 100
```

```python
df_table = df_bar.sort('2017_hover')
```

```python
output_file('raw_change.html')
source = ColumnDataSource(df_bar)

x = list(df_bar.city)
p = p = figure(plot_width = 800, 
               plot_height = 800, 
               x_range = (min(df_bar.change), max(df_bar.change)),
               y_range = x,
               title = 'Raw Change in Cvent Rankings from 2012 to 2017', 
               x_axis_label = 'Change from 2012 to 2017', 
               y_axis_label = 'City')

p.hbar(right = 'change', y = 'city', height = .7, left = 0, source = source)
# p.xaxis.major_label_orientation = math.pi/2
p.add_tools(HoverTool(tooltips = 
                     [
                         ("City", '@city'),
#                          ("Change in Cvent Ranking", "@change"),
                         ("2012", "@2012_hover"),
                         ('2013', '@2013_hover'),
                         ('2014', '@2014_hover'),
                         ('2015', '@2015_hover'),
                         ('2016', '@2016_hover'),
                         ('2017', '@2017_hover')
                     ]))

columns = [
        TableColumn(field="city", title="City"),
        TableColumn(field="2017_hover", title="2017 Cvent Ranking")]

table_source = ColumnDataSource(df_table)

data_table = DataTable(source = table_source, columns = columns)
data_table.row_headers = False
    
grid = gridplot([[p, data_table]])
    
show(grid)
save(grid)
```
