```python
import pandas as pd
import numpy as np
from bokeh.charts.attributes import cat
from bokeh.models import HoverTool
from bokeh.plotting import ColumnDataSource, figure, show, output_file, save
from bokeh.models.formatters import BasicTickFormatter
import bokeh.palettes
import math
```

```python
df_cvent = pd.read_excel('cvent_50_lists.xlsx')
```

```python
df_cvent.head()
```

```python
city_list = list(pd.concat([df_cvent[2012], df_cvent[2013], df_cvent[2014], df_cvent[2015], df_cvent[2016], df_cvent[2017]]).unique())
```

```python
for i in df_cvent[2012]:
    print i
```

```python
list_12 = []
list_13 = []
list_14 = []
list_15 = []
list_16 = []
list_17 = []
for c in city_list:
    df_c = df_cvent[df_cvent[2012] == c]
    if df_c.empty == False:
        df_c = df_c.reset_index()
        rank = df_c['index'].loc[0] + 1
        list_12.append({2012: rank,
                         'city': c})
    else:
        rank = 'unranked'
        list_12.append({2012: rank,
                        'city':c})
    df_2012 = pd.DataFrame(list_12)
    
    
    df_c = df_cvent[df_cvent[2013]== c]
    if df_c.empty == False:
        df_c = df_c.reset_index()
        rank = df_c['index'].loc[0] + 1
        list_13.append({2013: rank,
                         'city':c})
    else:
        rank = 'unranked'
        list_13.append({2013: rank,
                         'city': c})
    df_2013 = pd.DataFrame(list_13)
    
    
    df_c = df_cvent[df_cvent[2014]== c]
    if df_c.empty == False:
        df_c = df_c.reset_index()
        rank = df_c['index'].loc[0] + 1
        list_14.append({2014: rank,
                         'city':c})
    else:
        rank = 'unranked'
        list_14.append({2014: rank,
                         'city': c})
    df_2014 = pd.DataFrame(list_14)
    
    
    df_c = df_cvent[df_cvent[2015]== c]
    if df_c.empty == False:
        df_c = df_c.reset_index()
        rank = df_c['index'].loc[0] + 1
        list_15.append({2015: rank,
                         'city':c})
    else:
        rank = 'unranked'
        list_15.append({2015: rank,
                         'city': c})
    df_2015 = pd.DataFrame(list_15)
    
    
    df_c = df_cvent[df_cvent[2016]== c]
    if df_c.empty == False:
        df_c = df_c.reset_index()
        rank = df_c['index'].loc[0] + 1
        list_16.append({2016: rank,
                         'city':c})
    else:
        rank = 'unranked'
        list_16.append({2016: rank,
                         'city': c})
    df_2016 = pd.DataFrame(list_16)
    
    
    df_c = df_cvent[df_cvent[2017]== c]
    if df_c.empty == False:
        df_c = df_c.reset_index()
        rank = df_c['index'].loc[0] + 1
        list_17.append({2017: rank,
                         'city':c})
    else:
        rank = 'unranked'
        list_17.append({2017: rank,
                         'city': c})
    df_2017 = pd.DataFrame(list_17)
        
```

```python
df_ranks = df_2012.merge(df_2013, on = 'city').merge(df_2014, on = 'city').merge(df_2015, on = 'city').merge(df_2016, on = 'city').merge(df_2017, on = 'city')
```

```python
df_ranks = df_ranks.set_index('city')
```

```python
df_ranks = df_ranks.transpose()
```

```python
df_ranks = df_ranks.reset_index()
```

```python
df_ranks = df_ranks.rename(columns = {'index':'Year'})
```

```python
df_ranks['Washington, DC'] = [2, 7, 8, 6, 7, 6]
```

```python
df_ranks = df_ranks.drop('Washington, D.C.', axis = 1)
```

```python
df_ranks
```

```python
df_bar = df_ranks.transpose()
```

```python
df_bar.columns = df_bar.iloc[0]
```

```python
df_bar = df_bar.drop(['Year'])
```

```python
df_bar = df_bar.replace('unranked', 50)
```

```python
df_bar['change'] = df_bar[2012] - df_bar[2017]
```

```python
df_bar = df_bar.reset_index()
```

```python
colors = bokeh.palettes.d3['Category20'][20] + bokeh.palettes.d3['Category20'][20] + bokeh.palettes.d3['Category20'][20] + bokeh.palettes.d3['Category20'][20]
```

```python
output_file('rankings_bokeh.html')
source_city = ColumnDataSource(df_ranks)

p = figure(title = 'City Rankings', x_axis_label = 'Year', y_axis_label = 'Cvent Ranking',  y_range = (50,0))

for city, color in zip(reversed(df_ranks.columns[1:]), colors):
    line = p.line(x = 'Year', y = city, source = source_city, color = color)
    p.add_tools(HoverTool(tooltips=
                                [("City: ", "%s" % city),
                                 ("Rank", "@{%s}" % city),],
                                renderers=[line]))
save(p)
```

```python
for col in df_bar.columns:
    df_bar.rename(columns = {col: str(col)})
```

```python
df_bar.columns = ['city', '2012', '2013', '2014', '2015', '2016', '2017', 'change']
```

```python
df_bar = df_bar.sort('change', ascending = False)
```

```python
df_bar['percent_change'] = (df_bar['change'] / df_bar['2012']) * 100
```

```python
output_file('raw_change.html')
source = ColumnDataSource(df_bar)

x = list(df_bar.city)
p = p = figure(plot_width = 800, plot_height = 800, x_range = x, title = 'Raw Change in Cvent Rankings', x_axis_label = 'City', y_axis_label = 'Change')

p.vbar(x = 'city', top = 'change', width = .7, source = source)
p.xaxis.major_label_orientation = math.pi/2
p.add_tools(HoverTool(tooltips = 
                     [
                         ("City", '@city'),
                         ("Change in Cvent Ranking", "@change")
                     ]))
show(p)
save(p)
```

```python
df_percent = df_bar.sort('percent_change', ascending = False)
```

```python
output_file('percent_change.html')
source = ColumnDataSource(df_percent)

x = list(df_percent.city)
y = df_percent.percent_change
p = figure(plot_width = 800, 
           plot_height = 800, 
           x_range = x,
           y_range = (-275, 50),
           title = 'Percent Change in Rankings', 
           x_axis_label = 'City', 
           y_axis_label = 'Percent Change in Rankings')

p.vbar(x = 'city', top = 'percent_change', width = .7, source = source)
p.xaxis.major_label_orientation = math.pi/2
p.add_tools(HoverTool(tooltips = 
                     [
                         ("City", '@city'),
                         ("Percent Change", "@percent_change" + '%')
                     ]))
show(p)
save(p)
```
