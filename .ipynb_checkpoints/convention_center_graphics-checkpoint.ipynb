```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import math
from matplotlib.ticker import FormatStrFormatter
from bokeh.charts import Bar, output_file, show
from bokeh.charts.attributes import cat
from bokeh.models import HoverTool
from bokeh.plotting import ColumnDataSource
from bokeh.models.renderers import GlyphRenderer
from bokeh.plotting import figure
from bokeh.models.formatters import BasicTickFormatter
%matplotlib inline
```

```python
df_all = pd.read_csv('convention_center_datasets.csv', encoding = 'utf-8')
```

```python
df_all.info()
```

```python
df_ei = pd.concat([df_all['Annual Visitors'], df_all['Total Spending'], df_all['Name'], df_all['Total space'], df_all['Exhibition space']], axis = 1)
```

```python
df_ei = df_ei.rename(columns = {'Exhibition space':'exh_space', 'Annual Visitors':'attendance'})
```

```python
df_ei = df_ei[df_ei['Total Spending'] != '453,000,000 (KEC and KICC)']
```

```python
df_ei['Total Spending'] = df_ei['Total Spending'].astype(float)
```

```python
df_ei = df_ei.sort('Total Spending', ascending = False).reset_index(drop = True).fillna('Unknown')
```

```python
df_ei
```

```python
df_ei = df_ei[:21]
```

```python
plt.style.use('seaborn-colorblind')


fig, ax = plt.subplots(figsize = (12,9))
# ax.figure(figsize = (12,9))
plt.bar(df_ei.index, df_ei['Total Spending'])
ax.yaxis.set_major_formatter(FormatStrFormatter('%.0f'))
plt.xticks(df_ei.index, df_ei.Name, rotation = 90)
```

```python
df_ei = df_ei.reset_index()
```

```python
df_ei = df_ei.rename(columns = {'Total space': 'Space'})
```

```python
df_ei = df_ei.dropna()
```

```python
df_ei
```

```python
df_ei.info()
```

```python
df_ei = df_ei.rename(columns = {'Total Spending':'Spending'})
```

```python
df_ei.Spending = df_ei.Spending.astype
```

```python
from bokeh.models import Axis
from bokeh.plotting import output_file, save
```

```python
output_file('con_center_graph.html')
source = ColumnDataSource(data = dict(ei = df_ei.Spending,
                                     con_center = df_ei.Name,
                                     space = df_ei.Space,
                                     exh_space = df_ei.exh_space,
                                     attendance = df_ei.attendance))
hover = HoverTool(tooltips = [('Name', '@con_center'),
                              ('Total Floor Space', '@space'),
                              ('Exhibition Space', '@exh_space'),
                              ('Annual Attendance', '@attendance')])

x = list(df_ei.Name)
y = list(df_ei.Spending)
p = figure(
           x_range = x, 
#            y_range = range(0, max(df_ei.Spending)), 
           title = 'Economic Impact of Large US Convention Centers', 
           tools = [hover],
          )
p.vbar(x = x, top = y, width = 0.9, source = source)
p.xaxis.major_label_orientation = math.pi/3
p.xaxis.axis_label = 'Convention Center'
p.yaxis.axis_label = 'Economic Impact'
yaxis = p.select(dict(type=Axis, layout="left"))[0]
yaxis.formatter.use_scientific = False
show(p)
save(p)
```
